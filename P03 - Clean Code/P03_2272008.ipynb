````python
def main():
    n = int(input("N : "))

    # Segiempat (genap)
    if (n % 2 == 0):
        for i in range(n):
            for j in range(n):
                print("*", end=" ")
            print()

    # # Segitiga sama sisi(ganjil)
    else:
        for i in range(n):
            for j in range(i):
                print("*", end=" ")
            print()

if __name__ == "__main__":
    main()

````

````python
def input_buah():
    jenis_buah = str(input("Buah (Alpukat/Apel/Jeruk) : "))
    jumlah_buah = int(input("Jumlah {} (kg) : ".format(jenis_buah)))
    return jenis_buah, jumlah_buah

def hitung_total(jenis_buah, jumlah_buah, total):
    if jenis_buah == "Alpukat":
        total["alpukat"] += jumlah_buah
    elif jenis_buah == "Apel":
        total["apel"] += jumlah_buah
    elif jenis_buah == "Jeruk":
        total["jeruk"] += jumlah_buah

def print_total(total, total_kg, total_harga):
    print("=" * 35)
    print("Jumlah Buah Alpukat : ", total["alpukat"], "kg")
    print("Jumlah Buah Apel : ", total["apel"], "kg")
    print("Jumlah Buah Jeruk : ", total["jeruk"], "kg")
    print("Total Jumlah Buah yaitu", total_kg, "kg")
    print("Total Harga Buah Rp", total_harga)

def main():
    total = {"alpukat": 0, "apel": 0, "jeruk": 0}

    # Input dan hitung total untuk 3 jenis buah
    for i in range(3):
        jenis_buah, jumlah_buah = input_buah()
        hitung_total(jenis_buah, jumlah_buah, total)

    # Hitung total kg
    total_kg = sum(total.values())

    # Hitung total harga
    total_harga = total["apel"] * 35000 + total["jeruk"] * 20000 + \
    total["alpukat"] * 40000

    # Panggil fungsi print_total
    print_total(total, total_kg, total_harga)

if __name__ == "__main__":
    main()
````

> # **Kode Ketiga**














```
Before::

Kesalahan prinsip SOLID:
1. SRP -> Class Karyawan memiliki dua tanggung jawab (simpan data karyawan hitung gaji)
2. OCP -> Kode perlu diubah misalnya kita ingin menambah jenis karyawan baru dengan perhitungan gaji yang berbeda

```




````python
class Karyawan:
    def __init__(self, nama, gaji):
        self.nama = nama
        self.gaji = gaji

    def hitung_gaji(self):
        if self.gaji < 1000000:
            return self.gaji * 0.8
        else:
            return self.gaji * 0.9

def main():
    karyawan1 = Karyawan("Budi", 1500000)
    gaji_karyawan1 = karyawan1.hitung_gaji()

    print("Gaji karyawan1:", gaji_karyawan1)

if __name__ == "__main__":
    main()
````

````python
from abc import ABC, abstractmethod

class Karyawan(ABC):
    def __init__(self, nama):
        self.nama = nama

    @abstractmethod
    def hitung_gaji(self):
        pass

class KaryawanTetap(Karyawan):
    def __init__(self, nama, gaji):
        super().__init__(nama)
        self.gaji = gaji

    def hitung_gaji(self):
        return self.gaji

class KaryawanKontrak(Karyawan):
    def __init__(self, nama, gaji_per_jam, jam_kerja):
        super().__init__(nama)
        self.gaji_per_jam = gaji_per_jam
        self.jam_kerja = jam_kerja

    def hitung_gaji(self):
        return self.gaji_per_jam * self.jam_kerja

def main():
    karyawan1 = KaryawanTetap("Budi", 1500000)
    gaji_karyawan1 = karyawan1.hitung_gaji()
    print("Gaji karyawan1:", gaji_karyawan1)

    karyawan2 = KaryawanKontrak("Ani", 50000, 8)
    gaji_karyawan2 = karyawan2.hitung_gaji()
    print("Gaji karyawan2:", gaji_karyawan2)

if __name__ == "__main__":
    main()
````

> # **Kode Keempat**













```
Before::

Kesalahan prinsip SOLID:
1. DIP -> Kelas toko bergantung langsung pada kelas produk

```




````python
class Toko:
    def __init__(self, nama):
        self.nama = nama
        self.produk = []

    def tambah_produk(self, produk):
        self.produk.append(produk)

    def hapus_produk(self, produk):
        self.produk.remove(produk)

    def tampilkan_produk(self):
        for produk in self.produk:
            print(produk.nama, produk.harga)

class Produk:
    def __init__(self, nama, harga):
        self.nama = nama
        self.harga = harga


def main():
    toko1 = Toko("Toko Sembako")

    produk1 = Produk("Beras", 10000)
    produk2 = Produk("Gula", 8000)

    toko1.tambah_produk(produk1)
    toko1.tambah_produk(produk2)

    print('Kondisi awal: ')
    toko1.tampilkan_produk()
    print()
    toko1.hapus_produk(produk1)

    print('Setelah dihapus: ')
    toko1.tampilkan_produk()

if __name__ == "__main__":
    main()

````

```
After::
Membuat hubungan tengah (interface)
```


````python
from abc import ABC, abstractmethod

class IProduk(ABC):
    @abstractmethod
    def get_nama(self):
        pass

    @abstractmethod
    def get_harga(self):
        pass

class Produk(IProduk):
    def __init__(self, nama, harga):
        self.nama = nama
        self.harga = harga

    def get_nama(self):
        return self.nama

    def get_harga(self):
        return self.harga

class IToko(ABC):
    @abstractmethod
    def tambah_produk(self, produk):
        pass

    @abstractmethod
    def hapus_produk(self, produk):
        pass

    @abstractmethod
    def tampilkan_produk(self):
        pass

class Toko(IToko):
    def __init__(self, nama):
        self.nama = nama
        self.produk = []

    def tambah_produk(self, produk):
        self.produk.append(produk)

    def hapus_produk(self, produk):
        self.produk.remove(produk)

    def tampilkan_produk(self):
        for produk in self.produk:
            print(produk.get_nama(), produk.get_harga())

def main():
    toko1 = Toko("Toko Sembako")

    produk1 = Produk("Beras", 10000)
    produk2 = Produk("Gula", 8000)

    toko1.tambah_produk(produk1)
    toko1.tambah_produk(produk2)

    print('Kondisi awal: ')
    toko1.tampilkan_produk()
    print()
    toko1.hapus_produk(produk1)

    print('Setelah dihapus: ')
    toko1.tampilkan_produk()

if __name__ == "__main__":
    main()
````

````python

from abc import ABC, abstractmethod
class Hewan:
    def __init__(self, nama):
        self.nama = nama

    @abstractmethod
    def bersuara(self):
        pass

    @abstractmethod
    def berenang(self):
        pass

    @abstractmethod
    def terbang(self):
        pass

class Anjing(Hewan):
    def bersuara(self):
        print(f"{self.nama} menggonggong")

    def berenang(self):
        print(f"{self.nama} berenang")


class Burung(Hewan):
    def bersuara(self):
        print(f"{self.nama} berkicau")

    def terbang(self):
        print(f"{self.nama} terbang")

class Ikan(Hewan):
    def bersuara(self):
        print(f"{self.nama} tidak bersuara")

    def berenang(self):
        print(f"{self.nama} berenang")


def main():
    anjing1 = Anjing("Anjing")
    burung1 = Burung("Burung")
    ikan1 = Ikan("Ikan")

    anjing1.bersuara()
    anjing1.berenang()

    burung1.bersuara()
    burung1.terbang()

    ikan1.bersuara()
    ikan1.berenang()

if __name__ == '__main__':
    main()
````

````python
from abc import ABC, abstractmethod

class Hewan(ABC):
    @abstractmethod
    def bersuara(self):
        pass

class IBerjalan(ABC):
    @abstractmethod
    def berjalan(self):
        pass

class IBerenang(ABC):
    @abstractmethod
    def berenang(self):
        pass

class ITerbang(ABC):
    @abstractmethod
    def terbang(self):
        pass

class Anjing(Hewan, IBerjalan, IBerenang):
    def __init__(self, nama):
        self.nama = nama

    def bersuara(self):
        print(f"{self.nama} menggonggong")

    def berjalan(self):
        print(f"{self.nama} berjalan")

    def berenang(self):
        print(f"{self.nama} berenang")

class Burung(Hewan, ITerbang):
    def __init__(self, nama):
        self.nama = nama

    def bersuara(self):
        print(f"{self.nama} berkicau")

    def terbang(self):
        print(f"{self.nama} terbang")

class Ikan(Hewan, IBerenang):
    def __init__(self, nama):
        self.nama = nama

    def bersuara(self):
        print(f"{self.nama} tidak bersuara")

    def berenang(self):
        print(f"{self.nama} berenang")

def main():
    anjing1 = Anjing("Anjing")
    burung1 = Burung("Burung")
    ikan1 = Ikan("Ikan")

    anjing1.bersuara()
    anjing1.berenang()

    burung1.bersuara()
    burung1.terbang()

    ikan1.bersuara()
    ikan1.berenang()
if __name__ == '__main__':
    main()
````
